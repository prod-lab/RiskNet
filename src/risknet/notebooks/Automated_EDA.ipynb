## General Feature Data Profiles on Raw Data
### From Sampled FM 2009 Dataset

Using the ydata_profiling package to generate two interactive html files for each of the raw datasets from Freddie Mac. The Freddie Mac Single Family Loan-Level Dataset is split into two data files- the first being data on the original loan and the second being monthly performance data on loans. The two html files give basic data distribution metrics per feature. Opening these files with a live server allows for interactive exploration at the feature level for each dataset.

```python
#importing column names
from columns import orig_cols, monthly_cols

#reading in sample datasets
loan_df = pd.read_csv(
    fm_root+"sample_orig_2009.txt",
    sep='|',
    header=None,
    names= orig_cols,
    )

monthly_df = pd.read_csv(
    fm_root+"sample_svcg_2009.txt",
    sep='|',
    header=None,
    names = monthly_cols,
    )

loan_html = "loan_origination_data.html"
monthly_html = "monthly_perf_data.html"

if not (os.path.exists(loan_html) and os.path.exists(monthly_html)):

    loan_data_profile = ProfileReport(loan_df, title="Loan Origination Datafile Sample 2009")
    monthly_data_profile = ProfileReport(monthly_df, title="Monthly Performance Data Sample 2009")

    loan_data_profile.to_file(loan_html)
    monthly_data_profile.to_file(monthly_html)
else:
    print("HTML files already exist. Skipping profile creation.")
```

```python
#Global Imports:
import pandas as pd
from pandas import DataFrame

import numpy as np

from typing import List, Dict, Tuple
import pickle
import logging
logger = logging.getLogger("freelunch")

os.chdir('c:\\Users\\Zaki Ahmed\\Documents\\RiskNet\\src\\risknet\\')
sys.path.append(os.getcwd()+"\proc") #reorient directory to access proc .py files
import label_prep
import reducer
import encoder

#Pipeline:

#Step 1: Label Processing: Returns dev_labels.pkl and dev_reg_labels.pkl
label_prep.label_proc(fm_root, data)

#Step 2: Reducer: Returns df of combined data to encode
df = reducer.reduce(fm_root, data[0]) 
#As of right now, we are only pulling 2009 data. So we only need data[0].

#However, if we want to add 2014 data in the future, we can add another Tuple(str,str,str) to the List data
#and uncomment this code:
#df = reducer.reduce(fm_root, data[1])

#Data Cleaning 1: Define datatypes; Define what should be null (e.g., which codes per column indicate missing data)

 #Define datatypes
df = encoder.datatype(df)

#Define where it should be null:
#where we have explicit mappings of nulls
df = encoder.num_null(df)

# where we have explicit mappings of nulls
df = encoder.cat_null(df)

#Data Cleaning 2: Categorical, Ordinal, and RME Encoding
# interaction effects
df['seller_servicer_match'] = np.where(df.seller_name == df.servicer_name, 1, 0)

'''Categorical Encoding'''
df = encoder.cat_enc(df)

'''Ordinal Encoding'''
df = encoder.ord_enc(df, fm_root)

'''RME Encoding'''
df = encoder.rme(df, fm_root)
```

```python
pipeline_df_html = "proc_data.html"

if not (os.path.exists(pipeline_df_html)):

    df_profile = ProfileReport(df, title="Processed Data 2009Q1")

    df_profile.to_file(pipeline_df_html)
else:
    print("HTML file already exists. Skipping profile creation.")
```
